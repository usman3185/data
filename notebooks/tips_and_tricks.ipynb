Missing values and imputing them

```python
# find out the percentage of missing values

df.isnull().sum()/len(df) * 100
```

```python
# drop the deck column fom dataset  having more than 70% missing values#

df.drop('deck',axis=1, inplace=True)
```

```python
df['age'].mean()
```

```python
df['age'].median()
```

```python
df['age'].mode(0)
```

```python

# filling the missing values of age column by the mean of age

df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill embark and embarked town with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

# feature engineering using binning 

```python
#  binning af age columns into 7 categories

bins = [0,1,5,12,18,30,50,80]
labels = ['infants','toddler', 'kids', 'teens','young','middle age','old']
#  kis columns ko  bins ma convert krna
df['binned age']  =  pd.cut(df['age'], bins = bins , labels = labels)
```
