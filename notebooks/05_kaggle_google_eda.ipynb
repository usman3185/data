# check that how many MBs are there in one kb and how to calculate
1 convert kb into mbs 
2 then remove M form all numbers
3 handle 'varies with device'

```python


# Function to clean and convert Size column into MB
def convert_size(size):
    try:
        size = str(size).strip()

        # Ignore "Varies with device"
        if size == "Varies with device":
            return None

        # If size is already in MB
        if size.endswith("M"):
            return float(size.replace("M", ""))

        # If size is in KB → convert to MB
        if size.endswith("k"):
            kb_value = float(size.replace("k", ""))
            return kb_value / 1024  # KB to MB

        return None  # For unexpected cases
    except:
        return None

# Apply conversion to create new column
df["Size_MB"] = df["Size"].apply(convert_size)

# Display first 15 rows to verify
df[["App", "Size", "Size_MB"]].head(15)
```

```python


# Create the new column
df['Size_MB'] = df['Size'].str.extract(r'(\d+(?:\.\d+)?)', expand=False).astype(float)

# Apply the scaling conversion
df.loc[df['Size'].str.endswith('k'), 'Size_MB'] *= 1024
df.loc[df['Size'].str.endswith('M'), 'Size_MB'] *= 1024 ** 2
df.loc[df['Size'].str.endswith('G'), 'Size_MB'] *= 1024 ** 3

print(df)
```

```python


# Function to convert size to MB
def convert_size(size):
    if 'k' in size:
        return float(size.replace('k', '')) / 1024  # Convert KB to MB
    elif 'M' in size:
        return float(size.replace('M', ''))  # Already in MB
    else:
        return None  # Handle unexpected values

# Apply the conversion function to the 'Size' column
df['Size_MB'] = df['Size'].apply(convert_size)

print(df)
```
