```python
# convert the size column to numeric by multiplying the values with 1024 if it has 'K' in it and 1024*1024 if it has 'M' in it
def convert_size(size):
    """
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'K' in it and 1024*1024 if it has 'M' in it
    """
    if isinstance(size, str):
        if 'K' in size:
            return float(size.replace('K', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size

df['Size'] = df['Size'].apply(convert_size)
```

```python
df.head()


```

```python
# Replacing 'Varies with device' with NaN for better handling
df['Size']=df['Size'].replace('Varies with device', np.nan)

# Function to convert size in MB
def convert_size(size):
    if pd.isnull(size):
        return np.nan
    # remove commas before conversion
    size = size.replace(',', '')
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    return np.nan  # for any other cases, return NaN
# Apply the conversion function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)

# Check for missing values and datatype after conversion
print(f"Data type of 'Size' after conversion: {df['Size'].dtype}")
print(f"Missing values in 'Size' after conversion: {df['Size'].isnull().sum()}")
display(df[['Size']].head())
```

```python
df.describe().T
```

```python
# Rename the 'Size' column to 'App Size (MB)' for clarity
df.rename(columns={'Size': 'App Size (MB)'}, inplace=True)
```

```python
# Remove commas and '+' signs, then convert to numeric
df['Installs'] = df['Installs'].astype(str).str.replace('[+,]', '', regex=True)
df['Minimum Installs'] = df['Minimum Installs'].astype(str).str.replace('[+,]', '', regex=True)

# Convert to numeric (errors='coerce' will turn invalid values into NaN)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Minimum Installs'] = pd.to_numeric(df['Minimum Installs'], errors='coerce')
```

```python
df.head() # check the head of the dataframe
```

```python
df['Installs'].dtype # this will show the data type of the column
```

```python
# Fill missing values with median (better for skewed data like installs)
df['Installs'] = df['Installs'].fillna(df['Installs'].median())
df['Minimum Installs'] = df['Minimum Installs'].fillna(df['Minimum Installs'].median())

```

```python
# Check if missing values remain
print(df[['Installs', 'Minimum Installs']].isnull().sum()) 
```

```python
# Define bins (cut points) and labels
bins = [0, 1_000, 10_000, 100_000, 1_000_000, 10_000_000, float("inf")]
labels = ['0-1K', '1K-10K', '10K-100K', '100K-1M', '1M-10M', '10M+']

# Apply binning on Install column
df['Installs_Bin'] = pd.cut(df['Installs'], bins=bins, labels=labels, right=False) 

# Apply binning on Minimum Install column
df['MinInstalls_Bin'] = pd.cut(df['Minimum Installs'], bins=bins, labels=labels, right=False)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
# Convert 'Last Updated' and 'Scraped Time' to datetime objects
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Scraped Time'] = pd.to_datetime(df['Scraped Time'], errors='coerce')

# Check the data types after conversion
print("Data types after converting date columns:")
print(df[['Last Updated', 'Scraped Time']].dtypes)

# Check for any missing values introduced by conversion errors
print("\nMissing values after converting date columns:")
print(df[['Last Updated', 'Scraped Time']].isnull().sum())
```
