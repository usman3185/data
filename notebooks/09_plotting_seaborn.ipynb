```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
# df = your_dataframe

# Define custom colors for each hue level
custom_palette = {
    'male': 'blue',   # Color for males
    'female': 'pink'  # Color for females
}

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=custom_palette)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Sex')
plt.show()
# saving the plot 300dpi
plt.savefig('scatter_plot.png',dpi=300)
# saving the plot in vector file such as svg
plt.savefig('scatter_plot.svg')
```

```python
# style of plot
sns.set_theme(style='darkgrid',font_scale=1.2)
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
# df = your_dataframe

# Define custom colors for each hue level
custom_palette = {
    'male': 'blue',   # Color for males
    'female': 'pink'  # Color for females
}

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=custom_palette)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Sex')
plt.show()
```

# Boxplot

```python
# style of plot
sns.set_theme(style='darkgrid',font_scale=1.2)
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
# df = your_dataframe

# Define custom colors for each hue level
custom_palette = {
    'male': 'blue',   # Color for males
    'female': 'pink'  # Color for females
}

plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='sex', y='age', hue='sex', palette=custom_palette)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Sex')
plt.show()
```
